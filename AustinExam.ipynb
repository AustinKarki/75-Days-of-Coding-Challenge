```python
# process of oversampling
good=credit[credit["Class"]==1]
fraud=credit[credit["Class"]==0]
oversampling=fraud.sample(len(good))
new_credit=pd.concat([good,oversampling])
new_credit["Class"].value_counts()
```

```python
plt.figure(figsize=(15, 6)) # fixing the size as required
sns.countplot(x='Class', data=new_credit,palette='viridis') # Visualizing the good and fraud data in count plot
plt.title('Balanced Good And Fraud', fontsize=20, fontweight='bold', pad=25) # adding the padding in the topic
plt.xlabel('Class(1 or 0)',fontsize=16, fontweight='bold')
plt.ylabel('Count',fontsize=16, fontweight='bold')
plt.grid(True,linestyle="--") # adding a grid to understand more clearly about the data
plt.show()
```

```python
X=new_credit.drop("Class",axis=1) # dropping Class so that all other column can be included in features except Class
y=new_credit["Class"] # traget data
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=100)# spiliting into test 20 percent and train 80 percent as test size 0.2 given
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train) # fitting the data
```

```python
pred=model.predict(X_test) # prediction is done on X_test which is compared with y_test for finding accuracy
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred) # comparing the X_test prediction with y_test to find the mean_squared_error
print(f"MSE of the Dataset :{mse}")
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)
sns.heatmap(cm, annot=True,
            xticklabels=['Non-Fraud', 'Good'],
            yticklabels=['Non-Fraud', 'Good'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix',pad=25)
plt.show()
```

```python
from sklearn.metrics import classification_report
detail=classification_report(y_test,pred, target_names=['Non-Fraud', 'Good'])
print()
print("Confusion Matrix Report:\n\n")
print(detail)
```
